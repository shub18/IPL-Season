# <center> Team With Most IPL Title 

```python
win=matches[['season','winner']]
win
```

```python
team_with_most_title={}
for i in sorted(win.season.unique()):team_with_most_title[i]=win[win.season==i]['winner'].tail(1).values[0]
team_with_most_title

```
